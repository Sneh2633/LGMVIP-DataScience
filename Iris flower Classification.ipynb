## Analysis

```python
sns.heatmap(df.corr(), annot=True, cmap='Wistia')
```

```python
sns.pairplot(df)
```

```python
sns.catplot(x='petal length (cm)',y='petal width (cm)', palette='husl', hue='target', data=df)
```
